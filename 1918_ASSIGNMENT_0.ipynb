```python
import pandas as pd
data=pd.read_csv('/content/sample_data/california_housing_train.csv')
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
x=data.drop('median_house_value',axis=1)
y=data['median_house_value']
print(x)
print(y)
```

```python
from re import L
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
print('mean squared error',mse)
```

```python
r2=model.score(x_test,y_test)
print('R-squared',r2)
```

```python
from sklearn.tree import DecisionTreeRegressor

decision_tree_model = DecisionTreeRegressor(random_state=20)
decision_tree_model.fit(x_train, y_train)
```

```python
# Make predictions on the test set
y_pred_dt = decision_tree_model.predict(x_test)

# Calculate Mean Squared Error
mse_dt = mean_squared_error(y_test, y_pred_dt)
print("Decision Tree Mean Squared Error:", mse_dt)

# Calculate R-squared score
r2_dt = decision_tree_score = decision_tree_model.score(x_test, y_test)
print("Decision Tree R2 Score:", r2_dt)
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(x_train, y_train)
```

```python
y_pred_knn = knn_model.predict(x_test)

mse_knn = mean_squared_error(y_test, y_pred_knn)
print("KNN Mean Squared Error:", mse_knn)

r2_knn = knn_model.score(x_test, y_test)
print("KNN R2 Score:", r2_knn)
```

```python
df=pd.read_csv('/content/sample_data/mnist_train_small.csv', header=None)
print(df.head())
```

```python
y=df[0]
x=df.drop(0,axis=1)
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split

x_train_mnist, x_test_mnist, y_train_mnist, y_test_mnist = train_test_split(x, y, test_size=0.2, random_state=20)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

logistic_regression_model = LogisticRegression(max_iter=1000)
logistic_regression_model.fit(x_train_mnist, y_train_mnist)

y_pred_lr = logistic_regression_model.predict(x_test_mnist)

accuracy_lr = accuracy_score(y_test_mnist, y_pred_lr)
precision_lr = precision_score(y_test_mnist, y_pred_lr, average='weighted')
recall_lr = recall_score(y_test_mnist, y_pred_lr, average='weighted')
f1_lr = f1_score(y_test_mnist, y_pred_lr, average='weighted')

print("Logistic Regression Performance:")
print(f"Accuracy: {accuracy_lr:.4f}")
print(f"Precision: {precision_lr:.4f}")
print(f"Recall: {recall_lr:.4f}")
print(f"F1-score: {f1_lr:.4f}")
```

```python
from sklearn.svm import SVC

svm_model = SVC(kernel='linear')
svm_model.fit(x_train_mnist, y_train_mnist)

y_pred_svm = svm_model.predict(x_test_mnist)

accuracy_svm = accuracy_score(y_test_mnist, y_pred_svm)
precision_svm = precision_score(y_test_mnist, y_pred_svm, average='weighted')
recall_svm = recall_score(y_test_mnist, y_pred_svm, average='weighted')
f1_svm = f1_score(y_test_mnist, y_pred_svm, average='weighted')

print("\nSupport Vector Machine (SVM) Performance:")
print(f"Accuracy: {accuracy_svm:.4f}")
print(f"Precision: {precision_svm:.4f}")
print(f"Recall: {recall_svm:.4f}")
print(f"F1-score: {f1_svm:.4f}")
```

```python
from sklearn.ensemble import RandomForestClassifier

random_forest_model = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest_model.fit(x_train_mnist, y_train_mnist)

y_pred_rf = random_forest_model.predict(x_test_mnist)

accuracy_rf = accuracy_score(y_test_mnist, y_pred_rf)
precision_rf = precision_score(y_test_mnist, y_pred_rf, average='weighted')
recall_rf = recall_score(y_test_mnist, y_pred_rf, average='weighted')
f1_rf = f1_score(y_test_mnist, y_pred_rf, average='weighted')

print("\nRandom Forest Performance:")
print(f"Accuracy: {accuracy_rf:.4f}")
print(f"Precision: {precision_rf:.4f}")
print(f"Recall: {recall_rf:.4f}")
print(f"F1-score: {f1_rf:.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

models = ['Logistic Regression', 'SVM', 'Random Forest']
accuracy_scores = [accuracy_lr, accuracy_svm, accuracy_rf]

plt.figure(figsize=(8, 6))
plt.bar(models, accuracy_scores, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.ylabel('Accuracy')
plt.title('Comparison of Model Accuracy')
plt.ylim(0, 1)
plt.show()
```

```python
best_model_index = np.argmax(accuracy_scores)
best_model_name = models[best_model_index]
best_model_accuracy = accuracy_scores[best_model_index]

print(f"The model with the highest accuracy is: {best_model_name} with an accuracy of {best_model_accuracy:.4f}")
```
